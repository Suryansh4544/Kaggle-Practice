**Finally, ignoring the payments of 0 using the mask you created in the previous step, use pandas `.apply()` and NumPy's `np.log10()` to plot histograms of logarithmic transformations of the non-zero payments.**
> Hint: You can use `.apply()` to apply any function, including `log10`, to all the elements of a DataFrame or a column using the following syntax: `.apply(<function_name>)`.

```python
import numpy as np
df_1 = pay_amt_feats.replace(0,np.nan).dropna(axis=0).astype(float)
```

```python
df_1['pay_1']=df_1['PAY_AMT1'].apply(lambda x:np.log(x))
df_1['pay_2']=df_1['PAY_AMT2'].apply(lambda x:np.log(x))
df_1['pay_3']=df_1['PAY_AMT3'].apply(lambda x:np.log(x))
df_1['pay_4']=df_1['PAY_AMT4'].apply(lambda x:np.log(x))
df_1['pay_5']=df_1['PAY_AMT5'].apply(lambda x:np.log(x))
df_1['pay_6']=df_1['PAY_AMT6'].apply(lambda x:np.log(x))
df_1.dropna(inplace=True)
```

```python
import seaborn as sns
f, axes = plt.subplots(2, 3, figsize=(12, 7))
sns.distplot( df_1["pay_1"] , color="skyblue", ax=axes[0, 0],kde=False,bins=20)
sns.distplot( df_1["pay_2"] , color="black", ax=axes[0, 1],kde=False,bins=20)
sns.distplot( df_1["pay_3"] , color="gold", ax=axes[0, 2],kde=False,bins=20)
sns.distplot( df_1["pay_4"] , color="red", ax=axes[1, 0],kde=False,bins=20)
sns.distplot( df_1["pay_5"] , color="darkblue", ax=axes[1, 1],kde=False,bins=20)
sns.distplot( df_1["pay_5"] , color="crimson", ax=axes[1, 2],kde=False,bins=20)
```
